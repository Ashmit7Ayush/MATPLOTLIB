```python
import numpy as np
import matplotlib.pyplot as plt
```

 a  classifier  to  detect  the  Iris-Virginica  type  based  only  on  the  petalwidth feature

```python
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())
```

```python
# iris["data"],iris["target"],iris["target_names"]
iris["feature_names"]
```

```python
X = iris["data"][:,3:] #petal width
y = (iris["target"]==2).astype(np.int) # if iris virginiaca then 1 else 0
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X,y)
```

```python
X_new = np.linspace(0,3,1000).reshape(-1,1)# generating from 0 to 3 1000 number of samples
y_probs = log_reg.predict_proba(X_new)

```

```python
# y_probs,X_new
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_new,y_probs[:,1], "g-", label="iris-Virginica")
plt.plot(X_new,y_probs[:,0], "r--", label="Not - iris-Virginica")
plt.legend()
plt.xlabel('Petal width(cm)')
plt.ylabel('Probability')
plt.show()
```

```python
log_reg.predict([[1.7],[1.5]])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(iris["data"][(log_reg.predict(X)==1),2:3],iris["data"][(log_reg.predict(X)==1),3:],marker='D',label='IRIS_VERGINICA')

plt.scatter(iris["data"][(log_reg.predict(X)==0),2:3],iris["data"][(log_reg.predict(X)==0),3:],marker='^',label='not--IRIS_VERGINICA')

plt.xlabel('PETAL length')

plt.ylabel('PETAL width')

plt.legend()
plt.show()
```
