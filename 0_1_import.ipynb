MAtPlotLib Subplots

Display multiple plots

1 row 2 column
1st row 1st column
1st row 2st column


```python
# SubPlots()
plt.subplot(1,2,1)
plt.plot(x1,y1)

plt.subplot(1,2,2)
plt.plot(x2,y2)

plt.show()
```

2 row 1 col
1st row 1st col
2nd row 1st col

```python
plt.subplot(2,1,1)
plt.plot(x1,y1)

plt.subplot(2,1,2)
plt.plot(x2,y2,c='r',ms=10,ls='-',mec='y',mfc='g',lw='13',marker='P')

plt.show()
```

Assign as many as we want to provide

```python
plt.subplot(2,1,1)
plt.plot(x1,y1)
plt.title('test_!')

plt.subplot(2,1,2)
plt.plot(x2,y2)
plt.title('test_2')

plt.show()
```

```python
color = np.array([0,21,42,63,75])
plt.scatter(x2,y2,c=color,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
size = np.array([12,201,333,450,59])
```

```python
plt.scatter(x1,y1,c=color,s=size,cmap='viridis')
plt.colorbar()
plt.show()
```

MATPLOTLIB PIECHART

```python
plt.pie(y1)
plt.show()
```

```python
label=['ashmit','badal','bhaga','akash','gurnoor']
plt.pie(y1,labels=label)
plt.show()
```

we can also change the starting degree

```python
label=['ashmit','badal','bhaga','akash','gurnoor']
plt.pie(y1,labels=label,startangle=35)
plt.show()
```

explode makes the amout to come out at that extent

```python
label=['ashmit','badal','bhaga','akash','gurnoor']
plt.pie(y1,labels=label,startangle=35,explode=[0.1,0,0,0,0])
plt.show()
```

shadow

```python
label=['ashmit','badal','bhaga','akash','gurnoor']
plt.pie(y1,labels=label,startangle=35, explode=[0.1,0,0,0,0], shadow=True)
plt.show()
```

we can also give our own color as no of elements

startangle, shadow, explode, color, labels

plt.legend()   for the show up

```python
label=['ashmit','badal','bhaga','akash','gurnoor']
plt.pie(y1,labels=label,startangle=35, explode=[0.1,0,0,0,0], shadow=True)
plt.legend()
plt.show()
```

legend with header use of title

```python
label=['ashmit','badal','bhaga','akash','gurnoor']
plt.pie(y1,labels=label,startangle=35, explode=[0.1,0,0,0,0], shadow=True)
plt.legend(title='names')
plt.title('names')
plt.show()
```
